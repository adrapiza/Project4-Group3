```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
salaries_path = "data/Salaries.csv"
salaries = pd.read_csv(salaries_path)
salaries.head(10)
```

```python
salaries.describe()
```

```python
num_people = salaries['Id'].nunique()
num_people
```

```python
no_name_salaries = salaries.drop('EmployeeName', axis = 1)
no_name_salaries
```

```python
clean_salaries = no_name_salaries.set_index('Id')
clean_salaries
```

```python
filtered_clean_df = clean_salaries[~clean_salaries['JobTitle'].isin(['Not provided', 'Not Provided']) & 
                  ~clean_salaries['BasePay'].isin(['Not provided', 'Not Provided'])]

print("\nDataFrame after removing unwanted rows:")
filtered_clean_df
```

```python
filtered_clean_salaries = filtered_clean_df.drop('Notes', axis = 1)
filtered_clean_salaries
```

```python
num_title = filtered_clean_salaries['JobTitle'].nunique()
num_title
```

```python
ordered_titles = filtered_clean_salaries.sort_values(by='JobTitle')
ordered_titles
```

```python
ordered_totalPay = filtered_clean_salaries.sort_values(by='TotalPay', ascending = False)
ordered_totalPay
```

```python
relative_counts = ordered_titles['JobTitle'].value_counts(normalize=False)
relative_counts.head(50)
```
